```python
f = open('WhatsApp Chat with JioCare.txt', 'r', encoding='utf-8')
```

```python
print(data)
```

```python
pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
```

```python
messages = re.split(pattern, data)[2:]
messages
```
